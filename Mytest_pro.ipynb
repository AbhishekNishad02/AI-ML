```python
import pickle
import pandas as pd

# Load model
model = pickle.load(open("heart_failure_model.pkl", "rb"))

# Correct sample (M = male, F = female)
sample = pd.DataFrame([{
    "Age": 60,
    "Sex": "M",
    "ChestPainType": "ATA",
    "RestingBP": 130,
    "Cholesterol": 240,
    "FastingBS": 0,
    "RestingECG": "Normal",
    "MaxHR": 150,
    "ExerciseAngina": "N",
    "Oldpeak": 1.5,
    "ST_Slope": "Up"
}])

# Prediction
pred = model.predict(sample)
proba = model.predict_proba(sample)[:, 1]

print("Prediction:", pred)
print("Probability of Heart Disease:", proba)



```
